
## Import Data and dependencies

```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib

url = 'https://archive.ics.uci.edu/static/public/697/data.csv'
df1 = pd.read_csv(url)
```

## Spliting data into train and test

```python
X = df1.iloc[:, :-1]
Y = df1.iloc[:, -1]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=42)
```

## Import Decision Tree function and evaluation Predection

```python
model = DecisionTreeClassifier()
model = model.fit(X_train, Y_train)
prediction = model.predict(X_test)
prediction
```

## Validating confidence level

```python
confidence = model.predict_proba(X_test)
confidence
```

## Visalizing the tree

```python
matplotlib.pyplot.figure(figsize=(50,12))
tree.plot_tree(model)
```

## Result

```python
print(classification_report(Y_test, prediction))
```
